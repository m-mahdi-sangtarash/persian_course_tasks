2. create a python dict of anagrams, indexed by anagramed words.

```python
with open ('words.txt') as file:
    words = (file.readlines())

word_lst = []
    
for word in words:
    word_lst.append(word.strip().lower())
    
sort_data = sorted(set(word_lst)) #sort data in alphabet




def letter_sort (word):
    return ''.join(sorted(word))


word_dic = dict()

for word in sort_data:
    sig_word = letter_sort(word)
    if sig_word not in word_dic:
        word_dic[sig_word] = [word]
    
    else:
        word_dic[sig_word].append(word)
word_dic
```
